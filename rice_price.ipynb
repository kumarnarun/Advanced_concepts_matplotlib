```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Indian_rice_price',parse_dates=True)
df.head()
```

```python
df['date'] = df['date'].apply(pd.to_datetime)
df.set_index('date',inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
plt.plot(df['date'],df["price"])
plt.xlabel("Date")
plt.ylabel("price (rupees)")
plt.show()
```

```python
plt.scatter(df['date'], df['price'])
plt.title('Scatter Plot of Time Series Data')
plt.xlabel('Date')
plt.ylabel('price')
plt.show()
```

```python
df.boxplot(column=['price'])
plt.title('Box Plot of Time Series Data')
plt.ylabel('price')
plt.show()
```

```python
df['Month'] = df['date'].dt.month
df['Year'] = df['date'].dt.year
pivot_table = pd.pivot_table(df, values='price', index='Month', columns='Year')
plt.imshow(pivot_table, cmap='coolwarm', interpolation='nearest')
plt.title('Heatmap of Time Series Data')
plt.xlabel('Year')
plt.ylabel('Month')
plt.colorbar()
plt.show()
```

```python
df["year"] = df["date"].dt.year

# box plot grouped by year
sns.boxplot(df)
plt.show()
```
